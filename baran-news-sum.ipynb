```python
import pandas as pd

# Kaynak ve hedef dosyalarını belirtin
datasets = {
    "train": ("/content/drive/Shareddrives/Ilgar/dataset/MLSUM/tu/train.txt.src", "/content/drive/Shareddrives/Ilgar/dataset/MLSUM/tu/train.txt.tgt"),
    "val": ("/content/drive/Shareddrives/Ilgar/dataset/MLSUM/tu/val.txt.src", "/content/drive/Shareddrives/Ilgar/dataset/MLSUM/tu/val.txt.tgt"),
    "test": ("/content/drive/Shareddrives/Ilgar/dataset/MLSUM/tu/test.txt.src", "/content/drive/Shareddrives/Ilgar/dataset/MLSUM/tu/test.txt.tgt"),
}

# Her veri seti için işlem yap
for dataset_name, (src_path, tgt_path) in datasets.items():
    data = []

    # Kaynak ve hedef dosyalarını sırayla oku
    with open(src_path, 'r', encoding='utf-8') as src_file, open(tgt_path, 'r', encoding='utf-8') as tgt_file:
        src_lines = src_file.readlines()
        tgt_lines = tgt_file.readlines()

        # Kaynak ve hedef satırları eşleştirerek listeye ekle
        for src, tgt in zip(src_lines, tgt_lines):
            data.append({"Metin": src.strip(), "Özet": tgt.strip()})

    # Verileri bir DataFrame'e dönüştür
    df = pd.DataFrame(data)

    # CSV dosyasını kaydet
    csv_filename = f"/content/drive/Shareddrives/Ilgar/dataset/{dataset_name}.csv"
    df.to_csv(csv_filename, index=False, encoding='utf-8')
    print(f"{csv_filename} başarıyla oluşturuldu.")
```

```python
from datasets import Dataset
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, Seq2SeqTrainer, Seq2SeqTrainingArguments
import pandas as pd

# Hugging Face Dataset formatına dönüştür
train_dataset = Dataset.from_pandas(train)
val_dataset = Dataset.from_pandas(val)
test_dataset = Dataset.from_pandas(test)

# Tokenizer ve model
tokenizer = AutoTokenizer.from_pretrained("google/mt5-base")
model = AutoModelForSeq2SeqLM.from_pretrained("google/mt5-base")

# Preprocessing işlevi
def preprocess_function(examples):
    model_inputs = tokenizer(
        examples["Metin"],
        max_length=768,
        truncation=True,
        padding="max_length"
    )
    with tokenizer.as_target_tokenizer():
        labels = tokenizer(
            examples["Özet"],
            max_length=256,
            truncation=True,
            padding="max_length"
        )
    model_inputs["labels"] = labels["input_ids"]
    return model_inputs

# Tokenize edilmiş veri setleri
tokenized_train = train_dataset.map(preprocess_function, batched=True)
tokenized_val = val_dataset.map(preprocess_function, batched=True)
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
from evaluate import load
import pandas as pd

# Model ve tokenizer'ı yükle
load_directory = "/content/drive/Shareddrives/Ilgar/summarization_model"
tokenizer = AutoTokenizer.from_pretrained(load_directory)
model = AutoModelForSeq2SeqLM.from_pretrained(load_directory)

# Test veri setini yükle
test = pd.read_csv("/content/drive/Shareddrives/Ilgar/dataset/test_filtered.csv")
test = test[:1000]

# ROUGE metriğini yükle
rouge = load("rouge")

# Özetleme ve değerlendirme
references = []
predictions = []

for index, row in test.iterrows():
    # Test metnini al
    test_text = row["Metin"]
    reference_summary = row["Özet"]

    # Tokenizer ile girdiyi hazırla
    inputs = tokenizer(
        test_text,
        max_length=768,
        truncation=True,
        return_tensors="pt"
    )

    # Model ile özet üret
    outputs = model.generate(
        inputs["input_ids"],
        max_length=256,
        num_beams=4,
        early_stopping=True
    )

    # Üretilen özeti çöz
    generated_summary = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Referans ve tahminleri listeye ekle
    references.append(reference_summary)
    predictions.append(generated_summary)

# ROUGE skorlarını hesapla
result = rouge.compute(predictions=predictions, references=references, use_stemmer=True)

# ROUGE skorlarını yazdır
print("ROUGE-1:", result["rouge1"])
print("ROUGE-2:", result["rouge2"])
print("ROUGE-L:", result["rougeL"])
```

```python
# prompt: Metin özetleme için df'yi kullanarak Coverage hesapla.

import pandas as pd

# Dosya yolunu belirtin
file_path = "/content/drive/Shareddrives/Ilgar/test_with_predictions.csv"

df = pd.read_csv(file_path)

# Özet ve tahmin edilen özet sütunlarını alın
summaries = df["Özet"].astype(str).tolist()
predicted_summaries = df["Tahmin Edilen Özet"].astype(str).tolist()

# Coverage hesaplaması
coverage_scores = []
for summary, predicted_summary in zip(summaries, predicted_summaries):
  summary_words = set(summary.split())
  predicted_summary_words = set(predicted_summary.split())
  coverage = len(summary_words.intersection(predicted_summary_words)) / len(summary_words) if len(summary_words)>0 else 0
  coverage_scores.append(coverage)


# Coverage skorlarını DataFrame'e ekleyin
df["Coverage"] = coverage_scores

# DataFrame'i CSV dosyasına kaydedin
df.to_csv("/content/drive/Shareddrives/Ilgar/test_with_predictions_and_coverage.csv", index=False, encoding='utf-8')

print("Coverage skorları başarıyla hesaplandı ve CSV dosyasına kaydedildi.")

# İlk 10 satırı yazdır
print(df.head(10))

# Coverage'ın ortalamasını hesaplayın
average_coverage = df["Coverage"].mean()
print(f"\nOrtalama Coverage: {average_coverage}")
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
from evaluate import load
import pandas as pd

# Model ve tokenizer'ı yükle
load_directory = "/content/drive/Shareddrives/Ilgar/finetuned_model_best"
tokenizer = AutoTokenizer.from_pretrained(load_directory)
model = AutoModelForSeq2SeqLM.from_pretrained(load_directory)

# Test veri setini yükle
test = pd.read_csv("/content/drive/Shareddrives/Ilgar/dataset_2/test_filtered.csv")
test = test[:1000]

# ROUGE metriğini yükle
rouge = load("rouge")

# Özetleme ve değerlendirme
references = []
predictions = []

for index, row in test.iterrows():
    # Test metnini al
    test_text = row["content"]
    reference_summary = row["abstract"]

    # Tokenizer ile girdiyi hazırla
    inputs = tokenizer(
        test_text,
        max_length=768,
        truncation=True,
        return_tensors="pt"
    )

    # Model ile özet üret
    outputs = model.generate(
        inputs["input_ids"],
        max_length=256,
        num_beams=4,
        early_stopping=True
    )

    # Üretilen özeti çöz
    generated_summary = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Referans ve tahminleri listeye ekle
    references.append(reference_summary)
    predictions.append(generated_summary)

# ROUGE skorlarını hesapla
result = rouge.compute(predictions=predictions, references=references, use_stemmer=True)

# ROUGE skorlarını yazdır
print("ROUGE-1:", result["rouge1"])
print("ROUGE-2:", result["rouge2"])
print("ROUGE-L:", result["rougeL"])
```

```python
# Tahminleri yeni sütun olarak ekle
test["Tahmin Edilen Özet"] = predictions

# Yeni DataFrame'i CSV olarak kaydet
test.to_csv("/content/drive/Shareddrives/Ilgar/model2test_with_predictions.csv", index=False, encoding='utf-8')

print("Tahminler başarıyla eklendi ve CSV dosyası kaydedildi.")
```
